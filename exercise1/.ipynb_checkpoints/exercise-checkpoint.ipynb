## Question 1

In the lecture we used the nearest neighbour classifier to classify points from a toy dataset into either "red" or "blue" classes. We investigated how the performance changes as a function of model complexity and what this means for the performance of our classifier on unseen data. Instead of using a linear model as in the lecture, use a k-nearest neighbour model.

Use `make_blobs(n_samples=400, centers=23, random_state=42)` to create a simple dataset and use the `KNeighborsClassifier` classifier to answer the above questions.

```python
from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier

labels = ["b", "r"]
X, y = make_blobs(n_samples=400, centers=23, random_state=42)
y = np.take(labels, (y < 10))
```

```python
# Your solution
plt.scatter(X[:,0], X[:,1],c=y)
```

```python
from sklearn.model_selection import train_test_split

# we split the learning set into train and test sets by randomly picking entries, 
# so that all the features present in the learning set are represented in both
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)

plt.scatter(train_X[:,0], train_X[:,1],c=train_y, marker='^', label='train')
plt.scatter(test_X[:,0], test_X[:,1],c=test_y, marker='*', label='test')
plt.legend()
```

```python
# train the classifier
kNC = KNeighborsClassifier()
kNC.fit(train_X, train_y)

# evaluate the performance
score_train = kNC.score(train_X, train_y)
score_test = kNC.score(test_X, test_y)

print('performance on train set:', score_train)
print('performance on test set:', score_test)
```

```python
# check the performance when changing the train/test set ratio.
# I have done the loop over different random realisations of the division (changed random_state values)
# to answer the question, how does this affect the performance as I found that only one realisation was fluctuating.
def check_train_size(train_size, random_state):
    train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=train_size, random_state=random_state)
    kNC = KNeighborsClassifier()
    kNC.fit(train_X, train_y)

    score_train = kNC.score(train_X, train_y)
    score_test = kNC.score(test_X, test_y)
    return score_train, score_test

train_sizes = np.arange(1,10)*0.1

all_train_scores = []
all_test_scores = []
for random_state in range(0, 50):
    scores = [check_train_size(train_size, random_state) for train_size in train_sizes]
        
    scores = np.array(scores)
    all_train_scores.append(scores[:,0])
    all_test_scores.append(scores[:,1])
    plt.plot(train_sizes, scores[:,0], '*-g', alpha=0.1)
    plt.plot(train_sizes, scores[:,1], '^-y', alpha=0.1)
    

plt.plot(train_sizes, np.array(all_train_scores).mean(axis=0), '-g', linewidth=3, label='train set')
plt.plot(train_sizes, np.array(all_test_scores).mean(axis=0), '-y', linewidth=3, label='test set')
plt.xlabel('train set fraction')
plt.ylabel('accuracy')
plt.legend()
```

The tendency that can be seen is that when the train set is too small, the performance is worse, as not all the features are properly included and thus cannont be trained for. Then at around fraction = 0.5 the performance for test set reaches the plateau. I think the best way to split the data is at the point then the performance for the test set reaches the plato: we want to have enough data to do the training, but we still want to have enough data to do the proper testing of the model we used.

The performance for the test set is always worse than the performance for the train set - as during the training we try to account for all the features in the data, but we can't as our model is not an ideal representation of the data and/or we don't have enough data points for the training. So the model would be optimized better for the train set than for the test set.


```python
# test for the number of neighbours
def check_neighbors(n_neighbors, random_state):
    train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=random_state)
    kNC = KNeighborsClassifier(n_neighbors=n_neighbors)
    kNC.fit(train_X, train_y)

    score_train = kNC.score(train_X, train_y)
    score_test = kNC.score(test_X, test_y)
    return score_train, score_test

neighbors = np.arange(1,21)

all_train_scores = []
all_test_scores = []
for random_state in range(0, 50):
    scores = [check_neighbors(n_neighbors, random_state) for n_neighbors in neighbors]
        
    scores = np.array(scores)
    all_train_scores.append(scores[:,0])
    all_test_scores.append(scores[:,1])
    plt.plot(neighbors, scores[:,0], '*-g', alpha=0.1)
    plt.plot(neighbors, scores[:,1], '^-y', alpha=0.1)
    

plt.plot(neighbors, np.array(all_train_scores).mean(axis=0), '-g', linewidth=3, label='train set')
plt.plot(neighbors, np.array(all_test_scores).mean(axis=0), '-y', linewidth=3, label='test set')
plt.xlabel('neighbors')
plt.ylabel('accuracy')
plt.legend()
```

```python
# calculate the best value for the n_neighbors
MSE = np.array(all_test_scores).mean(axis=0)
ind = np.argmax(MSE)
minimum = neighbors[ind]
print(minimum)
```

Perfomance on the training dataset is higher that on the testing one. But both share the same features and tendencies: if one has a local maximum at point X, then other one will also have it; if one is decreasing, another does so as well.

KNeighbor Classifier with 4 neighbors is more complicated, as it will have more free parameters. 

For this dataset the best setting of `n_neighbors` is 7, as it corresponds to the maximum of the accuracy function for the test set.

## Question 2

This is a regression problem. It mostly follows the setup of the classification problem so you should be able to reuse some of your work.

* plot your dataset
* fit a kNN regressor with varying number of `n_neighbors` and compare each regressors predictions to the location of the training and testing points. 
* plot the mean squared error of the classifier as a function of `n_neighbors` for both training and testing datasets.
* comment on the similarities and differences between the performance on the testing and training dataset.
* find the best setting of `n_neighbors` for this dataset.
* why is this the best setting?
* can you explain why the mean square error on the training dataset plateaus between ~`n_neihgors`=5 to 15 at the value that it does?

Use `make_regression()` to create the dataset and use `KNeighborsRegressor` to answer the above questions. Take a look at scikit-learn's [`metrics`](http://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics) module to compute the mean squared error.

```python
def make_regression(n_samples=400, noise_level=0.8, random_state=2):
    rng = check_random_state(random_state)
    X = np.linspace(-2, 2, n_samples)
    y = 2 * X + np.sin(5 * X) + rng.randn(n_samples) * noise_level
    
    return X.reshape(-1, 1), y
```

```python
# Your solution
X, y = make_regression()
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)

plt.plot(train_X, train_y, '*', color='C4', label='train')
plt.plot(test_X, test_y, '^', color='orange', label='test')
plt.legend()
```

```python
from sklearn.neighbors import KNeighborsRegressor

# test for the number of neighbours
def check_neighbors(neighbors, train_X, train_y, X):
    kNR = KNeighborsRegressor(n_neighbors=neighbors)
    kNR.fit(train_X, train_y)

    predict = kNR.predict(X)
    return predict

neighbors = np.arange(1,21)
predict_train_y = [check_neighbors(n, train_X, train_y, train_X) for n in neighbors]
predict_test_y = [check_neighbors(n, train_X, train_y, test_X) for n in neighbors]

# compare each regressors predictions to the location of the training and testing points
plt.rcParams["figure.figsize"] = (12, 15)
f, axes = plt.subplots(5, 4, sharex=True, sharey=True)
for ind, ax in enumerate(axes.flatten()):
    ax.plot(train_y, predict_train_y[ind], '*', color='C4', label='train')
    ax.plot(test_y, predict_test_y[ind], '^', color='orange', label='test')
    ax.set_title('n_neighbors: {}'.format(ind+1))
axes.flatten()[0].legend()
f.text(0.5, 0.08, 'initial y', ha='center')
f.text(0.04, 0.5, 'predicted y', va='center', rotation='vertical')
f.subplots_adjust(wspace=0)
```

```python
# plot the mean squared error of the classifier as a function of 
# n_neighbors for both training and testing datasets.
plt.rcParams["figure.figsize"] = (8, 8)
from sklearn.metrics import mean_squared_error

neighbors = np.arange(1,40)
all_predict_train_y = []
all_predict_test_y = []
for random_state in range(0, 50):
    train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=random_state)
    predict_train_y = [mean_squared_error(train_y, check_neighbors(n, train_X, train_y, train_X)) for n in neighbors]
    predict_test_y = [mean_squared_error(test_y, check_neighbors(n, train_X, train_y, test_X)) for n in neighbors]

    all_predict_train_y.append(predict_train_y)
    all_predict_test_y.append(predict_test_y)
    
    plt.plot(neighbors, predict_train_y, '-*', color='C4', alpha=0.1)
    plt.plot(neighbors, predict_test_y, '-^', color='orange', alpha=0.1)
    
plt.plot(neighbors, np.array(all_predict_train_y).mean(axis=0), '-', color='C4', linewidth=3, label='train')
plt.plot(neighbors, np.array(all_predict_test_y).mean(axis=0), '-', color='orange', linewidth=3, label='test')
plt.xlabel('number of neighbors')
plt.ylabel('mean squared error')
plt.legend()
```

```python
# calculate the best value for the n_neighbors
MSE = np.array(all_predict_test_y).mean(axis=0)
ind = np.argmin(MSE)
minimum = neighbors[ind]
print(minimum)
```

The more complex model (small number of neighbors) performs very well on training set - it has enough free parameters to find a perfect mapping from X -> y. At the same time it performs worse for the test set, as it was overfitted on the train set. With descrease of freedom the error on train set grows, but the error on test set shrinks. At some point both reach a plateau. The plateau is reached as at some point the number of degrees of freedom  (DOF) is small enough so that there is no overfitting, but at the same time it allows for the good representation of the data. Passed surtain value, the number of DOF is too small to represent the data and the MSE curve of train set exits the plateau regime and starts to climb up.

The best setting of `n_neighbors` for this dataset should be at the location of the minimum for the mean square error of the test set, which is at `n_neighbor` = 18. 

The value at which the plateau is reached is dictated by the noise sigma, which is determined by the `noise_level` parameter in `make_regression` function.

Logistic regression. Use a linear model to solve a two class classification problem.

* What is the difference between a linear regression model and a logistic regression model?
* plot your data and split it into a training and test set
* draw your guess for where the decision boundary will be on the plot. Why did you pick this one?
* use the `LogisticRegression` classifier to fit a model to your training data
* extract the fitted coefficients from the model and draw the fitted decision boundary
* create a function to draw the decision surface (the classifier's prediction for every point in space)
* why is the boundary where it is?
* **(bonus)** create new datasets with increasingly larger amounts of noise (increase the `cluster_std` argument) and plot the decision boundary for each case. What happens and why?
* create 20 new datasets by changing the `random_state` parameter and fit a model to each. Visualise the variation in the fitted parameters and the decision boundaries you obtain. Is this a high or low variance model?

Use `make_two_blobs()` to create a simple dataset and use the `LogisticRegression` classifier to answer the above questions.

```python
from sklearn.linear_model import LogisticRegression

def make_two_blobs(n_samples=400, cluster_std=2., random_state=42):
    rng = check_random_state(random_state)
    X = rng.multivariate_normal([5,0], [[cluster_std**2, 0], [0., cluster_std**2]],
                                size=n_samples//2)
    
    X2 = rng.multivariate_normal([0, 5.], [[cluster_std**2, 0], [0., cluster_std**2]],
                                 size=n_samples//2)
    X = np.vstack((X, X2))
    return X, np.hstack((np.ones(n_samples//2), np.zeros(n_samples//2)))

X, y = make_two_blobs()
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))
```

The difference between a linear regression model and a logistic regression model is that linear regression is an actual regression i.e. it predicts a continous output, while logistic regression is used for classification purposes.

```python
# Your answer
# splitting data
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)

# predicting the division boundary
x_guess = np.linspace(np.min(X[:,0]), np.max(X[:,0]), 100)
y_guess = x_guess+0.15

plt.scatter(train_X[:,0], train_X[:,1], c=train_y, marker='^', label='train')
plt.scatter(test_X[:,0], test_X[:,1], c=test_y, marker='*', label='test')
plt.plot(x_guess, y_guess, '-', color='black', label='my guess')
plt.legend()
```

I have picked the boundary so that different types of points are maximally separated i.e. on the 'blue' side there is a minimum number of red points, and on the 'red' side there is a minimum number of the blue points.

```python
# fit model with logistic regression
lr = LogisticRegression()
lr.fit(train_X, train_y)

# extracted coeffs
coef = lr.coef_
intercept = lr.intercept_
print(coef, intercept)
# we derive the expression from the fact that at the boundary the predicted values are zero, 
# and the rest comes from the linear model
# 0 = intercept + coef1*x1 + coef2*x2
fitted_boundary = -(coef[0,0]*x_guess+intercept)/coef[0,1]

# plot
plt.scatter(train_X[:,0], train_X[:,1], c=train_y, marker='^', label='train')
plt.scatter(test_X[:,0], test_X[:,1], c=test_y, marker='*', label='test')
plt.plot(x_guess, y_guess, '--', color='black', label='my guess')
plt.plot(x_guess, fitted_boundary, '-', color='C4', label='log reg')
plt.legend()
```

```python
# drawing the decision surface
step = 0.1
x1min, x1max = np.min(X[:,0]), np.max(X[:,0])
x2min, x2max = np.min(X[:,1]), np.max(X[:,1])

xx1, xx2 = np.meshgrid(np.arange(x1min, x1max, step), np.arange(x2min, x2max, step))
yy = lr.predict(np.c_[xx1.ravel(), xx2.ravel()])
zz = [0.0 if el=='r' else 1.0 for el in yy]
zz = np.array(zz)
zz = zz.reshape(xx1.shape)

plt.pcolormesh(xx1, xx2, zz, cmap='RdBu')
plt.scatter(train_X[:,0], train_X[:,1], c=train_y, marker='^', label='train')
plt.scatter(test_X[:,0], test_X[:,1], c=test_y, marker='*', label='test')
plt.plot(x_guess, y_guess, '--', color='black', label='my guess')
plt.plot(x_guess, fitted_boundary, '-', color='C4', label='log reg')
plt.legend()
plt.xlim([x1min, x1max])
plt.ylim([x2min, x2max])
```

The difference is because I could see the whole dataset, but logistic regression saw only the training set, so this is what it was optimizing for. That's why the fitted boundary is there - it stills opt for minimumm 'intruder' point on the other's territory.

```python
# test for noise
def noisy_boundary(std):
    X, y = make_two_blobs(n_samples=400, cluster_std=std, random_state=42)
    labels = ['b', 'r']
    y = np.take(labels, (y < 0.5))
    
    train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)
    lr = LogisticRegression()
    lr.fit(train_X, train_y)
    coef = lr.coef_
    intercept = lr.intercept_

    return coef[0, 0], coef[0, 1], intercept

stds = np.arange(2, 10, 1)
for std in stds:
    coef0, coef1, intercpt = noisy_boundary(std)
    fitted_boundary = -(coef0*x_guess+intercpt)/coef1
    plt.plot(x_guess, fitted_boundary, '-', label='noise level: {}'.format(std))
    plt.legend()
```

The boundary slope gets steeper and steeper as more and more point leak to the other side.

```python
# Create 20 new datasets by changing the random_state parameter
def noisy_boundary(random):
    X, y = make_two_blobs(n_samples=400, cluster_std=2., random_state=random)
    labels = ['b', 'r']
    y = np.take(labels, (y < 0.5))
    
    train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)
    lr = LogisticRegression()
    lr.fit(train_X, train_y)
    coef = lr.coef_
    intercept = lr.intercept_

    return coef[0, 0], coef[0, 1], intercept

# Visualise the variation in the fitted parameters and the decision boundaries 
randoms = np.arange(10, 31, 1)
for random in randoms:
    coef0, coef1, intercpt = noisy_boundary(random)
    fitted_boundary = -(coef0*x_guess+intercpt)/coef1
    plt.plot(x_guess, fitted_boundary, '-')
    plt.title('decision boundaries')
    
```

```python
import itertools

params = [noisy_boundary(random) for random in randoms]
params = np.transpose(np.array(params))
titles = ['coef0', 'coef1', 'intercept']

plt.rcParams["figure.figsize"] = (10, 3)
f, axes = plt.subplots(1, 3, sharex=True)
for ax, param, title in zip(axes.flatten(), params, titles):
    ax.plot(randoms, param, '*')
    ax.set_title(title)

axes.flatten()[1].set_xlabel('random_state')
axes.flatten()[0].set_ylabel('param value')

stds = list(map(np.std, params))
print(stds)
```

Standard deviation of the parameters is of order of parameters themselves, so this is a high variance model.

## Question 4

Logistic regression. Use a more complex linear model to create a two class classifier for the "circle inside a circle" problem. Think about how you can increase the complexity of a logistic regression model. Visualise the classification accuracy as a function of the model complexity.

Use `make_circles(n_samples=400, factor=.3, noise=.1)` to create a simple dataset and use the `LogisticRegression` classifier to answer the above question.

```python
plt.rcParams["figure.figsize"] = (8, 8)
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=400, factor=.3, noise=.1)
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))

plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# Your answer
from sklearn.linear_model import LogisticRegression
# splitting data
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)

plt.scatter(train_X[:,0], train_X[:,1], c=train_y, marker='^', label='train')
plt.scatter(test_X[:,0], test_X[:,1], c=test_y, marker='*', label='test')
plt.legend()
```

To increase the complexity, I add some pseudo features to be trained for. I.E. instead of having just X[:,0]==x1 and X[:,1]==x2, I add the polynomial terms of these two variables: x1, x2, x1^2, x2^2, x1x2 etc. Like this I can get the circle-like features even with the linear regression.

```python
# adding additional features
# using complete homogeneous symmetric polynomial for this purpose
def extended_features(X, degree):
    X_augmented = []
    x1 = X[:,0]
    x2 = X[:,1]
    #x1 = np.array([1, 2, 3])
    #x2 = np.array([1, 1, 2])
        
    for i in range(1, degree + 1):
        for j in range(i+1):
            #print(i-j,j)
            xx = (x1 ** (i - j)) * (x2 ** j)
            X_augmented.append(xx)        
    return np.transpose(X_augmented)

# draw boundary line
def boundary(x1, x2, degree, coef, intercept):
    sum = np.zeros(shape=np.shape(x1))
    poly = []
    for i in range(1, degree + 1):
        for j in range(i+1):
            #print(i-j,j)
            xx = (x1 ** (i - j)) * (x2 ** j)
            poly.append(xx)
    poly = np.array(poly)
    coef = coef.reshape(-1,1)
    coef = np.array(coef)
    line = np.sum(np.multiply(poly,coef), axis=0)
    line = np.sum([line, intercept], axis=0)
    return line

# make plot
# unfortunately I didn't have time to find out
# how to properly label the contour plot for my purposes
def plot_boundary(degree, coef, intercept):
    step = 0.1
    x1min, x1max = np.min(X[:,0]), np.max(X[:,0])
    x2min, x2max = np.min(X[:,1]), np.max(X[:,1])
    x1, x2 = np.meshgrid(np.arange(x1min, x1max, step), np.arange(x2min, x2max, step))

    line = boundary(x1.ravel(), x2.ravel(), degree, coef, intercept)
    line = line.reshape(x1.shape)

    plt.scatter(train_X[:,0], train_X[:,1], c=train_y, marker='^', label='train')
    plt.scatter(test_X[:,0], test_X[:,1], c=test_y, marker='*', label='test')
    CS = plt.contour(x1, x2, line, [0])
    #plt.clabel(CS, inline=1, fontsize=10)
    #plt.legend()

# all features together
def vary_complexity(degree):
    # create extended features for the classifier
    train_XX = extended_features(train_X, degree)
    # train the classifier
    lr = LogisticRegression()
    lr.fit(train_XX, train_y)
    coef = lr.coef_[0]
    intercept = lr.intercept_
    # plot the result
    plot_boundary(degree, coef, intercept)
    

degrees = np.arange(1, 5)
for degree in degrees:
    vary_complexity(degree)
```

```python
# accuracy as a function of complexity
def accuracy_vs_complexity(degree):
    # create extended features for the classifier
    train_XX = extended_features(train_X, degree)
    test_XX = extended_features(test_X, degree)
    # train the classifier
    lr = LogisticRegression()
    lr.fit(train_XX, train_y)
    train_score = lr.score(train_XX, train_y)
    test_score = lr.score(test_XX, test_y)

    return train_score, test_score

degrees = np.arange(1, 10)
train_scores = [accuracy_vs_complexity(degree)[0] for degree in degrees]
test_scores = [accuracy_vs_complexity(degree)[1] for degree in degrees]
    
plt.plot(degrees, train_scores, '*', label='train')
plt.plot(degrees, test_scores, '^', label='test')
plt.xlabel('model complexity: degree of polynomial')
plt.ylabel('accuracy')
plt.legend()
```
