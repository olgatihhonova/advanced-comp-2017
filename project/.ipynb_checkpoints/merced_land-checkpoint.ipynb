# First we test how the problem behaves for simple 2 class classification. We choose planes and agricultural as classes.

#### First we try simple model

```python
# initial params
train_data_dir = 'UCMerced_LandUse/Images2D/train'
validation_data_dir = 'UCMerced_LandUse/Images2D/validation'
test_data_dir = 'UCMerced_LandUse/Images2D/test'
img_width, img_height = 256, 256
nb_train_samples = 60*2
nb_validation_samples = 20*2
epochs = 10
batch_size = 20
```

```python
# test our data augmentation 
datagen = ImageDataGenerator(
        samplewise_center=True,
        samplewise_std_normalization=True,
        rescale=1./255,
        rotation_range=180,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')

#i=0
#for batch in datagen.flow(x, batch_size=1,
#                          save_to_dir='UCMerced_LandUse/test', save_prefix='plane', save_format='jpeg'):
#    i += 1
#    if i > 5:
#        break  # otherwise the generator would loop indefinitely
```

```python
img = load_img('UCMerced_LandUse/Images2D/train/airplane/airplane00.bmp')
x = img_to_array(img)  # this is a Numpy array with shape (256, 256, 3)
x = x.reshape((1,) + x.shape)  # this is a Numpy array with shape (1, 256, 256, 3)

plt.imshow(img)
plt.title('original image');
```

```python
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
batches = datagen.flow(x, batch_size=1)
for ax, batch in zip(axes.flatten(), batches):
    batch = array_to_img(batch[0])
    ax.imshow(batch)
    ax.set_yticklabels([])
    ax.set_xticklabels([])
axes[0][1].set_title('augmented images');
```

```python
# the augmentation configuration for training
train_datagen = ImageDataGenerator(
        samplewise_center=True,
        samplewise_std_normalization=True,
        rescale=1./255,
        rotation_range=180,
        width_shift_range=0.2,
        height_shift_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')

# the augmentation configuration for validation with just rescaling
validation_datagen = ImageDataGenerator(
        samplewise_center=True,
        samplewise_std_normalization=True,
        rescale=1./255)

# # the augmentation configuration for testing with just rescaling
# test_datagen = ImageDataGenerator(
#         samplewise_center=True,
#         samplewise_std_normalization=True,
#         rescale=1./255)
```

```python
# initialise the generators
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

validation_generator = validation_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

# test_generator = test_datagen.flow_from_directory(
#     test_data_dir,
#     target_size=(img_width, img_height),
#     batch_size=batch_size,
#     class_mode='binary')
```

```python
# prepare the model
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=(img_width, img_height, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.summary()
```

```python
# compile the model
model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])
```

```python
# fit the model
history = model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size)
```

```python
print('minimum loss:', np.min(history.history['val_loss']))
print('maximum accuracy:', np.max(history.history['val_acc']),)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
ax0, ax1 = axes[0], axes[1]
ax0.plot(history.history['loss'], label='test loss')
ax0.plot(history.history['val_loss'], label='validation loss')
ax0.set_xlabel('epochs')
ax0.set_ylabel('loss')
ax0.legend(loc='best')

ax1.plot(history.history['acc'], label='test loss')
ax1.plot(history.history['val_acc'], label='validation loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('accuracy')
ax1.legend(loc='best');
```

```python
top_model_weights_path = 'bottleneck_fc_model_2D.h5'
```

```python
def save_bottlebeck_features():
    # build the VGG16 network
    model = applications.VGG16(include_top=False, weights='imagenet')
    
    datagen = ImageDataGenerator(rescale=1. / 255)
    generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)    
    bottleneck_features_train = model.predict_generator(
        generator, nb_train_samples // batch_size)
    np.save(open('bottleneck_features_train_2D.npy', 'wb'),
            bottleneck_features_train)

    generator = datagen.flow_from_directory(
        validation_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)   
    bottleneck_features_validation = model.predict_generator(
        generator, nb_validation_samples // batch_size)
    np.save(open('bottleneck_features_validation_2D.npy', 'wb'),
            bottleneck_features_validation)
```

```python
#save_bottlebeck_features()
```

```python
def train_top_model():
    train_data = np.load(open('bottleneck_features_train_2D.npy', 'rb'))
    train_labels = np.array(
        [0] * int(nb_train_samples / 2) + [1] * int(nb_train_samples / 2))

    validation_data = np.load(open('bottleneck_features_validation_2D.npy', 'rb'))
    validation_labels = np.array(
        [0] * int(nb_validation_samples / 2) + [1] * int(nb_validation_samples / 2))

    model = Sequential()
    model.add(Flatten(input_shape=train_data.shape[1:]))
    model.add(Dense(256, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(optimizer='rmsprop',
                  loss='binary_crossentropy', metrics=['accuracy'])

    history = model.fit(train_data, train_labels,
              epochs=epochs,
              batch_size=batch_size,
              validation_data=(validation_data, validation_labels))
    model.save_weights(top_model_weights_path)
    
    return history
```

```python
history = train_top_model()
```

```python
print('minimum loss:', np.min(history.history['val_loss']))
print('maximum accuracy:', np.max(history.history['val_acc']),)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
ax0, ax1 = axes[0], axes[1]
ax0.plot(history.history['loss'], label='test loss')
ax0.plot(history.history['val_loss'], label='validation loss')
ax0.set_xlabel('epochs')
ax0.set_ylabel('loss')
ax0.legend(loc='best')

ax1.plot(history.history['acc'], label='test loss')
ax1.plot(history.history['val_acc'], label='validation loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('accuracy')
ax1.legend(loc='best');
```

Let's plot confusion matrix.

```python
plt.rcParams["figure.figsize"] = (14,14)
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=65)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

# Now let's try different approach.
# Leverage VGG16 model features. 
We first run the convolution part of the VGG16 model once on our data to update the features, and them train the fully connected part through multiple epochs.

It the original blog they used it, as cats and dogs are aldready containd in the ImageNet dataset, on which the VGG16 model was trained. However, some of the classes of our current problem are also presented in the ImageNet. So, let's see what happens.

```python
top_model_weights_path = 'bottleneck_fc_model.h5'
```

```python
def save_bottlebeck_features():
    # build the VGG16 network
    model = applications.VGG16(include_top=False, weights='imagenet')
    
    datagen = ImageDataGenerator(
        samplewise_center=True,
        samplewise_std_normalization=True,
        rescale=1./255)
    generator = datagen.flow_from_directory(
        train_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)    
    bottleneck_features_train = model.predict_generator(
        generator, nb_train_samples // batch_size)
    np.save(open('bottleneck_features_train.npy', 'wb'),
            bottleneck_features_train)

    generator = datagen.flow_from_directory(
        validation_data_dir,
        target_size=(img_width, img_height),
        batch_size=batch_size,
        class_mode=None,
        shuffle=False)   
    bottleneck_features_validation = model.predict_generator(
        generator, nb_validation_samples // batch_size)
    np.save(open('bottleneck_features_validation.npy', 'wb'),
            bottleneck_features_validation)
```

```python
#save_bottlebeck_features()
```

```python
train_data = np.load(open('bottleneck_features_train.npy', 'rb'))
length = int(nb_train_samples/nb_classes)
train_labels = np.array([ [label]*length for label in np.arange(nb_classes) ]).flatten()

validation_data = np.load(open('bottleneck_features_validation.npy', 'rb'))
length = int(nb_validation_samples/nb_classes)
validation_labels = np.array([ [label]*length for label in np.arange(nb_classes) ]).flatten()
```

```python
# now lets shuffle the data
ind = np.arange(len(train_data))
np.random.shuffle(ind)
train_data_shuffle = train_data[ind]
train_labels_shuffle = train_labels[ind]

ind = np.arange(len(validation_data))
np.random.shuffle(ind)
validation_data_shuffle = validation_data[ind]
validation_labels_shuffle = validation_labels[ind]
```

```python
model = Sequential()

model.add(Flatten(input_shape=train_data.shape[1:]))

model.add(Dense(512, activation='relu',
                kernel_initializer='random_uniform',
                bias_initializer='zeros')) #, 
                #kernel_regularizer=regularizers.l2(0.001) ))
model.add(BatchNormalization())
model.add(Dropout(0.7))

model.add(Dense(512, activation='relu',
                kernel_initializer='random_uniform',
                bias_initializer='zeros')) #,
                #kernel_regularizer=regularizers.l2(0.1) ))
model.add(BatchNormalization())
model.add(Dropout(0.7))


model.add(Dense(21, activation='softmax',
                kernel_initializer='random_uniform',
                bias_initializer='zeros'))

model.summary()
```

```python
#opt = optimizers.Adagrad(lr=0.1, decay=0)
#opt = optimizers.Adam(lr=0.1)
#opt = optimizers.RMSprop(lr=0.001)
#opt = 'rmsprop'
opt = optimizers.SGD(lr=0.1, momentum=0.9, decay=1e-6, nesterov=True)
#opt = optimizers.SGD(lr=0.1, momentum=0.9, decay=0, nesterov=True)
#opt = optimizers.Nadam(lr=0.002, beta_1=0.9, beta_2=0.999, epsilon=1e-08, schedule_decay=0.004)

model.compile(loss='sparse_categorical_crossentropy',
              optimizer=opt, 
              metrics=['accuracy'])
```

```python
epochs = 50
batch_size = 100
```

```python
history = model.fit(train_data_shuffle, train_labels_shuffle,
          epochs=epochs,
          batch_size=batch_size,
          validation_data=(validation_data_shuffle, validation_labels_shuffle))

model.save_weights(top_model_weights_path)
```

```python
print('minimum loss:', np.min(history.history['val_loss']))
print('maximum accuracy:', np.max(history.history['val_acc']),)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
ax0, ax1 = axes[0], axes[1]
ax0.plot(history.history['loss'], label='test loss')
ax0.plot(history.history['val_loss'], label='validation loss')
ax0.set_xlabel('epochs')
ax0.set_ylabel('loss')
ax0.legend(loc='best')

ax1.plot(history.history['acc'], label='test accuracy')
ax1.plot(history.history['val_acc'], label='validation accuracy')
ax1.set_xlabel('epochs')
ax1.set_ylabel('accuracy')
ax1.legend(loc='best');
```

# Summary
I have tried several options for the flattened part of the model. I have tried Dense layers in combinations 64x64, 128x128, 256x256, 512x512, 1024x1024, 4096x4096, 512x256x64, 512x512x256, 512x512x256x256. With Dropout rates np.arange(0.5, 1, 0.1) after first and second, or only after first layer. With and without BatchNormalization. With and without kernel weight and bias initialization. With and without l2 norm. The best results I could achieve were validation accuracy of 60%, while train accuracy around 97%. So I was always facing an overfitting problem. To fight with it, I tried augmenting dropout rate, putting batch normalization and l2 norm. Unfortunately this didn't help, and almost everytime independent of model combination I had validation accuracy around 50% and train accuracy around 90%. I tried mostly with 20-50 epochs, but the more promosing models were computed during 200-300 epochs.

I have also tried different optimizers: RMSprop, Adam, Nadam, SGD. I was playing with learning rate depending on differennt models, usually in range 0.0001 to 0.1. 

I have read two articles on the problem https://arxiv.org/pdf/1508.00092.pdf and http://www.repository.up.ac.za/dspace/bitstream/handle/2263/51310/Luus_Multiview_2015.pdf?sequence=1. I was hoping to reproduce the results from the first paper, as they were also using the feature leverage from already pretrained model. Although they were using CAffeNet and I used VGG16, the differences between these two models are minor. But even using the flattened architecture as the VGG16 one with Dense4096+Dropout0.5+Dense4096+Dropout0.5+Dense21, that they were using in the article, I couldn't achieve higher validation accuracy than 60%. The difference might come form the use of different optimizer or some additional step not described in the paper. 

# Let's try GridSearch for the last fully connected model.
We search for the optimal l2 params and dropout rate.

```python
def make_model(alpha1, alpha2, dropout_rate1, dropout_rate2):
    model = Sequential()

    model.add(Flatten(input_shape=train_data.shape[1:]))

    model.add(Dense(512, activation='relu', 
                    kernel_initializer='random_uniform',
                    bias_initializer='zeros', 
                    kernel_regularizer=regularizers.l2(alpha1) ))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate1))
    
    model.add(Dense(512, activation='relu', 
                    kernel_initializer='random_uniform',
                    bias_initializer='zeros',
                    kernel_regularizer=regularizers.l2(alpha2) ))
    model.add(BatchNormalization())
    model.add(Dropout(dropout_rate2))

    model.add(Dense(21, activation='softmax',
                    kernel_initializer='random_uniform',
                    bias_initializer='zeros'))

    model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop', 
              metrics=['accuracy'])

    return model
```

```python
my_classifier = KerasClassifier(make_model, batch_size=batch_size)
validator = GridSearchCV(my_classifier,
                         param_grid={'alpha1': [0, 0.00001, 0.0001, 0.001, 0.01, 0.1],
                                     'alpha2': [0, 0.00001, 0.0001, 0.001, 0.01, 0.1],
                                     'dropout_rate1': np.arange(0.4, 0.9, 0.1),
                                     'dropout_rate2': np.arange(0.4, 0.9, 0.1),
                                     'epochs': [30]},
                        scoring='neg_log_loss',
                        cv=4,
                        n_jobs=1,
                        verbose=100)
```
