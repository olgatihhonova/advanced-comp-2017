```python
%config InlineBackend.figure_format='retina'
%matplotlib inline

import numpy as np
np.random.seed(123)
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (8, 8)
plt.rcParams["font.size"] = 14
from sklearn.utils import check_random_state
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
# test for the change of the coordinates
def transform_coordinates(X):
    Z = np.zeros(shape=X.shape)
    for idx,coord in enumerate(X):
        x = coord[0]
        y = coord[1]
        rho = np.sqrt(x**2 + y**2)
        theta = np.arctan2(y,x)
        Z[idx] = [rho, theta]
    return Z

# produce dataset
X, y = make_circles(n_samples=400, factor=.3, noise=.1)
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))

# transform the dataset
train_X, test_X, train_y, test_y = train_test_split(X, y, train_size=0.5, random_state=42)
Z = transform_coordinates(X)
train_Z = transform_coordinates(train_X)
test_Z = transform_coordinates(test_X)

# apply the model
lr = LogisticRegression()
lr.fit(train_Z, train_y)
coef = lr.coef_
intercept = lr.intercept_
print(coef, intercept)

# do the plotting
step = 0.005
x1min, x1max = np.min(Z[:,0]), np.max(Z[:,0])
x2min, x2max = np.min(Z[:,1]), np.max(Z[:,1])

x_guess = np.linspace(x1min, x1max, 100)
fitted_boundary = -(coef[0,0]*x_guess+intercept)/coef[0,1]

xx1, xx2 = np.meshgrid(np.arange(x1min, x1max, step), np.arange(x2min, x2max, step))
yy = lr.predict(np.c_[xx1.ravel(), xx2.ravel()])
zz = [0.0 if el=='r' else 1.0 for el in yy]
zz = np.array(zz)
zz = zz.reshape(xx1.shape)

plt.pcolormesh(xx1, xx2, zz, cmap='RdBu')
plt.scatter(train_Z[:,0], train_Z[:,1], c=train_y, marker='^', label='train')
plt.scatter(test_Z[:,0], test_Z[:,1], c=test_y, marker='*', label='test')
plt.plot(x_guess, fitted_boundary, '-', color='C4', label='boundary')
plt.legend()
plt.xlim([x1min, x1max])
plt.ylim([x2min, x2max])
```
