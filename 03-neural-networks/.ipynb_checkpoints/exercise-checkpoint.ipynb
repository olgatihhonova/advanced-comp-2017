## Question 1

Finding analytic gradients of expressions. Use the backprop framework we built in the lecture to compute the analytic gradient of an expression. This is useful to get thinking about
writing computations as graphs and to see first hand that there is no magic involved
in automatically finding derivatives.

For example, using the expression: $f(x) = \sin(x^2)$: implement a module for $\sin(x)$, build the graph representing the function, plot the expression as well as the gradient as a function of $x$.

* add a new operation (e.g. $\sin, \cos, \exp, x^y$, ...) to the `Addition` and `Multiply` modules.
* build a new expression using the available expressions
* plot your expression as well as its gradient
* compare the gradient to one you worked out by hand (or some other software package)

ps. if you want to use automatic differentiation for serious projects outside of neural
network projects checkout packages like https://github.com/HIPS/autograd

```python
class Multiply:
    def forward(self, x, y):
        self.x = x
        self.y = y
        return x * y
    
    def backward(self, dLdz):
        dzdx = self.y
        dLdx = dLdz * dzdx
        
        dzdy = self.x
        dLdy = dLdz * dzdy
        return [dLdx, dLdy]


class Add:
    def forward(self, x, y):
        self.x = x
        self.y = y
        return x + y
        
    def backward(self, dLdz):
        dzdy = 1
        dzdx = 1
        return [dLdz * dzdy, dLdz * dzdx]
    
    
class Sinus:
    def forward(self, x):
        self.x = x
        return np.sin(x)
    
    def backward(self, dLdz):
        dzdx = np.cos(self.x)
        dLdx = dLdz * dzdx
        return dLdx
    
    
class Square:
    def forward(self, x):
        self.x = x
        return x * x
    
    def backward(self, dLdz):
        dzdx = 2 * self.x
        dLdx = dLdz * dzdx
        return dLdx
```

```python
def f_with_gradients(x, y, z):
    q = Add()
    f = Multiply()
    
    q_out = q.forward(x, y)
    f_out = f.forward(q_out, z)
    
    grad_f = f.backward(1.)
    grad_q = q.backward(grad_f[0])

    gradients = [grad_q[0], grad_q[1], grad_f[1]]
    return f_out, gradients
```

```python
# your solution here
# function to calculate sin(x^2) and its grads
def f1_with_gradients(x):
    sq = Square()
    sin = Sinus()
    
    sq_out = sq.forward(x)
    sin_out = sin.forward(sq_out)
    
    grad_sin = sin.backward(1.0)
    grad_sq = sq.backward(grad_sin)
    
    return sin_out, grad_sq
```

```python
# plotting
x = np.arange(-2.5, 2.5, 0.1)

# the actual function and gradient for the check
sinx2 = np.sin(x*x)
grad = 2*x * np.cos(x*x)

f, f_grad = f1_with_gradients(x)

plt.plot(x, f, label='sinx2')
plt.plot(x, f_grad, label='backward grad')
plt.plot(x, sinx2, '*', label='reference func')
plt.plot(x, grad, '*', label='reference grad')
plt.legend()
```

## Question 2

Use the circle data set from exercise 1 and build a neural network classifier that can solve the problem (scikit-learn provides a MLPClassifier classifier that implements a neural network). Comment on:

* what is the minimum width of each layer

```python
# your answer
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
plt.rcParams["figure.figsize"] = (8, 8)

X, y = make_circles(n_samples=1000, factor=.3, noise=.1)
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))

plt.scatter(X[:,0], X[:,1], c=y)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.5, random_state=42)
```

```python
# let's search for best params using GridSearch
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPClassifier
plt.rcParams["figure.figsize"] = (8, 8)

parameter_grid = [{'activation': ['relu'], 'hidden_layer_sizes': np.arange(1, 5)},
                  {'activation': ['tanh'], 'hidden_layer_sizes': np.arange(1, 5)}]

mlpc = MLPClassifier(hidden_layer_sizes=(1), max_iter=3200,
                    random_state=21, activation='relu')
clf = GridSearchCV(mlpc, parameter_grid, cv=5)
clf.fit(X_train, y_train)
print(clf.best_score_, clf.best_params_, 'best score')

means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
paramss = clf.cv_results_['params']
for mean, std, params in zip(means, stds, paramss):
    print("%0.3f (+/-%0.03f) for %r" % (mean, std, params))
```

```python
from utils import plot_surface
plt.rcParams["figure.figsize"] = (8, 4)
# train with relu
clf_r = MLPClassifier(hidden_layer_sizes=(3), max_iter=3200, 
                    random_state=21, activation='relu')
clf_r.fit(X_train, y_train)

clf_t = MLPClassifier(hidden_layer_sizes=(3), max_iter=3200, 
                    random_state=21, activation='tanh')
clf_t.fit(X_train, y_train)
# plot
fig, axes = plt.subplots(1,2, sharey=True)
clfs = [clf_r, clf_t]
names = ['relu', 'tanh']
for ax, clf, title in zip(axes.flatten(), clfs, names):
    #plot_surface(clf, X_test, y_test, ax=ax)
    xlim = X[:, 0].min(), X[:, 0].max()
    ylim = X[:, 1].min(), X[:, 1].max()
    n_steps=250
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], n_steps),
                         np.linspace(ylim[0], ylim[1], n_steps))
    z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
    #z = np.array([0.0 if el=='r' else 1.0 for el in z])
    z = z.reshape(xx.shape)
    ax.contourf(xx, yy, z, alpha=0.8, cmap=plt.cm.RdBu_r)
    ax.scatter(X[:, 0], X[:, 1], c=y, s=10)
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_title(title)
print(clf_r.score(X_train, y_train), 'relu', clf_t.score(X_train, y_train), 'tanh')
```

```python
fig, axes = plt.subplots(1,2)
clfs = [clf_r, clf_t]
names = ['relu', 'tanh']
for ax, clf, title in zip(axes.flatten(), clfs, names):
    ax.plot(clf.loss_curve_)
    ax.set_xlabel('iterations')
    ax.set_ylabel('loss function')
    ax.set_title(title)
    ax.set_ylim([0, 1])
```

```python
# difference between tanh and relu
plt.rcParams["figure.figsize"] = (8, 8)
ind_relu = [i for i, p in enumerate(paramss) if p['activation']=='relu']
ind_tanh = [i for i, p in enumerate(paramss) if p['activation']=='tanh']
print(means[ind_relu], 'relu')
print(means[ind_tanh], 'tanh')
plt.plot(means[ind_relu], means[ind_tanh], '*')
plt.plot(means[ind_relu], means[ind_relu], '-')
plt.xlabel('relu')
plt.ylabel('tanh')
```

```python
# with polynomial features
plt.rcParams["figure.figsize"] = (8, 4)
X, y = make_circles(n_samples=1000, factor=.3, noise=.1)
labels = ['b', 'r']
y = np.take(labels, (y < 0.5))
# add features x1^2, x2^2
x1, x2 = X[:,0], X[:,1]
XX = np.transpose([x1, x2, np.square(x1), np.square(x2)])
X_train, X_test, y_train, y_test = train_test_split(XX, y, train_size=0.5, random_state=42)


clf_r = MLPClassifier(hidden_layer_sizes=(1), max_iter=3200, 
                    random_state=20, activation='relu')
clf_r.fit(X_train, y_train)
print(clf_r.score(X_train, y_train), 'relu')

clf_t = MLPClassifier(hidden_layer_sizes=(1), max_iter=3200, 
                    random_state=21, activation='tanh')
clf_t.fit(X_train, y_train)
print(clf_t.score(X_train, y_train), 'tanh')

fig, axes = plt.subplots(1,2, sharey=True)
clfs = [clf_r, clf_t]
names = ['relu', 'tanh']
for ax, clf, title in zip(axes.flatten(), clfs, names):
    #plot_surface(clf, X_test, y_test, ax=ax)
    xlim = X[:, 0].min(), X[:, 0].max()
    ylim = X[:, 1].min(), X[:, 1].max()
    n_steps=250
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], n_steps),
                         np.linspace(ylim[0], ylim[1], n_steps))
    z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel(), np.square(xx.ravel()), np.square(yy.ravel())])[:, 1]
    #z = np.array([0.0 if el=='r' else 1.0 for el in z])
    z = z.reshape(xx.shape)
    ax.contourf(xx, yy, z, alpha=0.8, cmap=plt.cm.RdBu_r)
    ax.scatter(X[:, 0], X[:, 1], c=y, s=10)
    ax.set_xlim(*xlim)
    ax.set_ylim(*ylim)
    ax.set_title(title)
print(clf_r.score(X_train, y_train), 'relu', clf_t.score(X_train, y_train), 'tanh')
```

## Question 2.5

Use the spiral data set build a neural network classifier that can solve the problem (scikit-learn provides a MLPClassifier classifier that implements a neural network). Comment on:
* what is the minimum number of layers
* what is the minimum width of each layer
* is there a difference between using the tanh and ReLU activation functions?
* (bonus) build the same neural network using `keras` instead.
* (bonus) can you create a visualisation to show how the network transforms the problem? (Tim isn't sure this can be done, so don't spend forever on this.)

```python
def make_spiral():
    N = 100 # number of points per class
    K = 3 # number of classes
    X = np.zeros((N*K, 2)) # data matrix (each row = single example)
    y = np.zeros(N*K, dtype='uint8') # class labels
    for j in range(K):
        ix = range(N*j, N*(j+1))
        r = np.linspace(0.0, 1, N) # radius
        t = np.linspace(j*4, (j+1)*4, N) + np.random.randn(N)*0.2 # theta
        X[ix] = np.c_[r*np.sin(t), r*np.cos(t)]
        y[ix] = j

    return X, y
```
